```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/Video Games Sales.csv")
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.dropna(subset=['Year'], inplace=True)
df.dropna(subset=['Publisher'], inplace=True)
```

```python
df.isnull().sum()
```

```python
sorted_df = df.sort_values(by='Global', ascending=False)

top_games = sorted_df.head(10)

print(top_games[['Game Title', 'Global']])
```

```python
plt.figure(figsize=(12, 6))
plt.barh(top_games['Game Title'], top_games['Global'], color='skyblue')
plt.xlabel('Global Sales (in millions)')
plt.title('Top Best-Selling Video Games')

plt.gca().invert_yaxis()

plt.show()
```
