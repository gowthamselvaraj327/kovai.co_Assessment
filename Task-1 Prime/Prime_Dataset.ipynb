```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/prime.csv")
```

```python
df['director'].fillna('Unknown', inplace=True)
```

```python
df['cast'].fillna('Unknown', inplace=True)
```

```python
df['country'].fillna('Unknown', inplace=True)
```

```python
df.dropna(subset=['date_added'], inplace=True)
```

```python
mode = df['rating'].mode().values[0]
df['rating']= df['rating'].replace(np.nan, mode)
```

```python
mode = df['duration'].mode().values[0]
df['duration']= df['duration'].replace(np.nan, mode)
```

```python
df.isnull().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()


df['cast'] = le.fit_transform\
                    (df['cast'])

df['duration'] = le.fit_transform\
                    (df['duration'])
```

```python
sns.scatterplot( x="type", y='rating', data=df, hue='type')


plt.legend(bbox_to_anchor=(1, 1), loc=2)

plt.show()
```

```python
sns.boxplot(x='duration', y="rating", data=df)
plt.show()
```

```python
sns.scatterplot(x='cast', y="rating", data=df)
plt.legend(bbox_to_anchor=(1, 1), loc=2)
plt.show()
```
